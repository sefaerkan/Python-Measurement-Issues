```python
# Derecelendirme, Satın Alma, Yoruma Göre Sıralama ( Sorting b)
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
df = pd.read_csv("product_sorting.csv")
```

```python
# Satın alma sayılarını 1 ile 5 arasındaki sayılara çevirdik.
df["purchase_count_scaled"] = MinMaxScaler(feature_range=(1,5)). \
fit(df[["purchase_count"]]).transform(df[["purchase_count"]])
```

```python
# Yorum sayılarını 1 ile 5 arasındaki sayılara çevirdik.
df["comment_count_scaled"] = MinMaxScaler(feature_range=(1,5)). \
fit(df[["commment_count"]]).transform(df[["commment_count"]])
```

```python
# Ayrıntı yaparak oran vererek skor'ları aldık.
# Satın almaya %26, yorumlara %32, puana ise %42 ağırlık verdik.
(df["comment_count_scaled"] * 32 / 100 +
 df["purchase_count_scaled"] * 26 / 100 +
 df["rating"] * 42 / 100)
```

```python
# Fonksiyon haline getirdik
def weighted_sorting_score(df,w1=32,w2=26,w3=42):
    return (df["comment_count_scaled"] * w1 / 100 +
            df["purchase_count_scaled"] * w2 / 100 +
            df["rating"] * w3 / 100)
```

```python
# Çıkan skorları yeni bir değişken oluşturup atadık.
df["weighted_sorting_score"] = weighted_sorting_score(df)
```

```python
# Çıkan değerimize göre sıraladık.
df.sort_values("weighted_sorting_score",ascending=False).head(20)
```

```python
# Bonus isminde sadece veri bilimi olanları görüntülemek istiyoruz.
df[df["course_name"].str.contains("Veri Bilimi")].sort_values("weighted_sorting_score",ascending=False).head(20)
```
